Machine Learning models can not deal with missing values. So we need to find a way to get rid of them.  
1. Assign a smaller dataset to a new variable by dropping the columns with more than 500 missing values. 
**Hint : Use** [pandas.DataFrame.drop](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html) 
2. In the folder where the data was located there is a text file named `data_description.txt`, open it and check the meaning of these variables!
3. Drop all the rows with missing Garage,MasVnr,Bsmt,Electrical related missing values!
**Hint : Use** [pandas.DataFrame.dropna](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html) 
4. fill the missing values in the column LotFrontage with the average of the column. 
**Hint : Use** [pandas.DataFrame.fillna](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html) 

```python
sum_null=raw_df.isna().sum()
a=sum_null[sum_null>500]
raw_df=raw_df.drop(a.index,axis=1)
raw_df.head()
```

```python
raw_df=raw_df.dropna(subset=['GarageType','MasVnrType','BsmtQual','Electrical'])

raw_df
```

```python
avg=raw_df.LotFrontage.mean()
raw_df.LotFrontage=raw_df.LotFrontage.fillna(avg)
raw_df
```
