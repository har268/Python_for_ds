## Introduction

Data can be understood better when we visualized them. Matplotlib is a flexible library which can help you to visualize your data.

In order to start to Data Visualization: 
- make the necessary imports, 
- prepare some data, 
- you can start plotting with the help of the `plot()` function!,
- Once you created your plot, you might need to explicitly write a command to show it, `show()`.

````python
import numpy as np
import matplotlib.pyplot as plt
````

Let's start by creating a Numpy Array.

````python
x = np.array([0, 10, 5, 7, 2, 1, 8, 8, 4, 4])

plt.plot(x,'r')
plt.show()
````

Note that this results in a figure that contains the values (y axis) and their correspondign index number (x axis)

````python
### Scatter Plot
````

````python
y = [0, 10, 5, 7, 2, 1, 8, 8, 4, 4]
x = [.1, .2, .3, .4, .5, .6, .7, .8, .9, 1] 

plt.scatter(x, y,c='r')
#plt.show()
````

````python
for c in ['r', 'g', 'b']:
    x = np.random.rand(1000)
    y = np.random.rand(1000)
    plt.scatter(x, y, color=c, s=np.random.randint(0, 100, size=(x.shape)), alpha=0.3)
````

````python
x = [1, 2, 3]
y = [6, 7, 9]

plt.scatter(x, y, s=[300, 300, 300], label='Hours of Sleep');
plt.title('Hours of Sleep in  the Last three Days')

plt.xlabel('X-Axis', fontsize=20)
plt.ylabel('Y-Axis', fontsize=20)

plt.xlim(1, 4)
plt.ylim(5, 10)

plt.xticks([1, 2, 3], ['Day 1', 'Day 2', 'Day 3'], fontsize=20)
plt.yticks([5, 7, 9], ['5 Hours', '7 Hours', '9 Hours'], fontsize=20)

plt.grid()
plt.legend(loc="upper left", fontsize=10);
````

### Anatomy of matplotlib plot

In essence, there are two big components that you need to take into account:

- The **Figure** is the overall window or page that everything is drawn on. Its the top-level component of all the ones that you will consider in the following points. You can create multiple independent Figures. A Figure can have several other things in it, such as a **subtitle**, which is a centered title to the figure. Youll also find that you can add a **legend** and **color bar**, for example, to your Figure.

- To the figure you add Axes. The Axes is the area on which the data is plotted with functions such as `plot()` and `scatter()` and that can have ticks, labels, etc. associated with it. This explains why Figures can contain multiple Axes.

Pretty much whatever we did above can be done with the axis object as well. In other words, the methods used to manipulate the properties of a plot, for example `plt.xlabel()` all exist for an axes object as well. For instacne, instead of `plt.xlabel()` we can also use `ax.set_xlabel()`.  

### Subplots

````python
fig, axes = plt.subplots(nrows=2, ncols=1)
````

````python
axes
````

````python
x = np.arange(0, 360, .01)
x_rad = np.deg2rad(x)
y1 = np.sin(x_rad)
y2 = np.cos(x_rad)
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[1].plot(x, y1)
axes[0].plot(x, y2);
````

````python
(ax1, ax2) = axes
````

````python
x = np.arange(0, 360, .01)
x_rad = np.deg2rad(x)
y1 = np.sin(x_rad)
y2 = np.cos(x_rad)
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
ax1.plot(x, y1)
ax2.plot(x, y2)
````

````python
fig, axes = plt.subplots(nrows=3, ncols=3)
for ax in axes.flat:
    ax.plot(np.random.random(100))
````

````python
fig, axes = plt.subplots(nrows=3, ncols=3)
for ax in axes.flat:
    ax.plot(np.random.random(100))
fig.tight_layout()
````

````python
fig, axes = plt.subplots(nrows=3, ncols=3, sharex=True, sharey=True)
for ax in axes.flat:
    ax.plot(np.random.random(100))
````
